```python
#CALIDAD
from datetime import date
class Proveedor:
    def __init__(self, id_proveedor, nombre):
        self.__id_proveedor=id_proveedor
        self.__nombre=nombre
    def get_id_proveedor(self):
        return self.__id_proveedor
    def set_id_proveedor(self, valor):
        self.__id_proveedor=valor
    def get_nombre(self):
        return self.__nombre
    def set_nombre(self, valor):
        self.__nombre=valor
class Almacen:
    def __init__(self, id_almacen, nombre):
        self.__id_almacen=id_almacen
        self.__nombre=nombre
    def get_id_almacen(self):
        return self.__id_almacen
    def set_id_almacen(self, valor):
        self.__id_almacen=valor
    def get_nombre(self):
        return self.__nombre
    def set_nombre(self, valor):
        self.__nombre=valor
class Producto:
    def __init__(self, id_producto, nombre):
        self.__id_producto=id_producto
        self.__nombre=nombre
    def get_id_producto(self):
        return self.__id_producto
    def set_id_producto(self, valor):
        self.__id_producto=valor
    def get_nombre(self):
        return self.__nombre
    def set_nombre(self, valor):
        self.__nombre=valor
ESTADOS_INGRESO=[e.lower() for e in ["Recibido", "En boca de inspección", "En inspección", "Autorizado para venta", "Producto devuelto a proveedor debido a fallas", "Retornado al inventario por parte de proveedor"]]
ESTADOS_TRANSFERENCIA=[e.lower() for e in ["Preparado para transferencia", "En tránsito hacia sucursal", "Recibido en sucursal"]]
ESTADOS_EGRESO=[e.lower() for e in ["En bandeja", "Recogido de la bandeja", "Empacado para enviar", "Enviado a cliente"]]
ESTADOS_DEVOLUCION=[e.lower() for e in ["Devuelto por cliente", "Evaluación de devolución del cliente", "Producto reincorporado al inventario"]]
ESTADOS_ACTUALES=ESTADOS_INGRESO + ESTADOS_TRANSFERENCIA + ESTADOS_EGRESO + ESTADOS_DEVOLUCION
ESTADOS_FINALES=["pendiente", "cerrado", "rechazado"]
ESTADOS_VALIDOS={
    "recibido": ["pendiente"],
    "en boca de inspección": ["pendiente"],
    "en inspección": ["pendiente"],
    "autorizado para venta": ["cerrado"],
    "producto devuelto a proveedor debido a fallas": ["pendiente", "rechazado"],
    "retornado al inventario por parte de proveedor": ["cerrado"],
    "preparado para transferencia": ["pendiente"],
    "en tránsito hacia sucursal": ["pendiente"],
    "recibido en sucursal": ["cerrado"],
    "en bandeja": ["pendiente"],
    "recogido de la bandeja": ["pendiente"],
    "empacado para enviar": ["pendiente"],
    "enviado a cliente": ["cerrado"],
    "devuelto por cliente": ["pendiente"],
    "evaluación de devolución del cliente": ["pendiente"],
    "producto reincorporado al inventario": ["cerrado"]
}
class Calidad:
    def __init__(self, id_inspeccion, proveedor, producto, almacen, estado_actual, estado_final, fecha_estado_actual: date, cantidad_devuelta:int=0, observaciones:str=""):
        estado_actual=estado_actual.lower()
        estado_final=estado_final.lower()
        if estado_actual not in ESTADOS_ACTUALES:
            raise ValueError(f"Estado actual '{estado_actual}' no es válido.")
        if estado_final not in ESTADOS_FINALES:
            raise ValueError(f"Estado final '{estado_final}' no es válido.")
        estados_finales_permitidos=ESTADOS_VALIDOS.get(estado_actual,[])
        if estado_final not in estados_finales_permitidos:
            raise ValueError(f"Estado final '{estado_final}' no es válido para el estado actual '{estado_actual}'.")
        self.id_inspeccion=id_inspeccion
        self.proveedor=proveedor
        self.producto=producto
        self.almacen=almacen
        self.estado_actual=estado_actual
        self.estado_final=estado_final
        self.fecha_estado_actual=fecha_estado_actual
        self.cantidad_devuelta=cantidad_devuelta
        self.observaciones=observaciones
        self.fecha_reingreso_almacen=None
        self.fecha_devolucion_proveedor=None
        if self.estado_actual in ["retornado al inventario por parte de proveedor", "producto reincorporado al inventario"]:
            self.fecha_reingreso_almacen=fecha_estado_actual
        elif self.estado_actual=="producto devuelto a proveedor debido a fallas":
            self.fecha_devolucion_proveedor=fecha_estado_actual
        self.historial_estados=[]
        self.guardar_historial(self.estado_actual, self.estado_final, self.fecha_estado_actual)
    def guardar_historial(self, estado_actual, estado_final, fecha):
      self.historial_estados.append({
            "estado_actual":estado_actual,
            "estado_final":estado_final,
            "fecha":fecha
      })
    def cambiar_estado(self, nuevo_estado_actual, nueva_fecha, nuevo_estado_final=None):
      nuevo_estado_actual=nuevo_estado_actual.lower()
      if nuevo_estado_actual not in ESTADOS_ACTUALES:
          raise ValueError(f"Nuevo estado actual '{nuevo_estado_actual}' no es válido.")
      if nuevo_estado_final:
          nuevo_estado_final=nuevo_estado_final.lower()
      else:
          nuevo_estado_final=self.estado_final
      if nuevo_estado_final not in ESTADOS_FINALES:
          raise ValueError(f"Nuevo estado final '{nuevo_estado_final}' no es válido.")
      estados_finales_permitidos=ESTADOS_VALIDOS.get(nuevo_estado_actual, [])
      if nuevo_estado_final not in estados_finales_permitidos:
          raise ValueError(f"Estado final '{nuevo_estado_final}' no es válido para el estado actual '{nuevo_estado_actual}'.")
      self.estado_actual=nuevo_estado_actual
      self.fecha_estado_actual=nueva_fecha
      self.estado_final=nuevo_estado_final
      if self.estado_actual in ["retornado al inventario por parte de proveedor","producto reincorporado al inventario"]:
          self.fecha_reingreso_almacen=nueva_fecha
      else:
          self.fecha_reingreso_almacen=None
      if self.estado_actual=="producto devuelto a proveedor debido a fallas":
          self.fecha_devolucion_proveedor=nueva_fecha
      else:
          self.fecha_devolucion_proveedor=None
      self.guardar_historial(self.estado_actual, self.estado_final, self.fecha_estado_actual)
    def determinar_accion(self):
        if self.estado_actual=="producto devuelto a proveedor debido a fallas":
            return "Devolver a proveedor"
        elif self.estado_actual in ["producto reincorporado al inventario", "retornado al inventario por parte de proveedor"]:
            return "Reingreso a inventario"
        elif self.estado_actual in ESTADOS_TRANSFERENCIA:
            return "Transferencia a sucursal"
        elif self.estado_actual in ESTADOS_EGRESO:
            return "Salida al cliente"
        elif self.estado_actual in ESTADOS_DEVOLUCION:
            return "Devolución de cliente"
        else:
            return "Proceso de inspección o ingreso"
    def mostrar_resultados(self):
      resultado=(
          f"--- INSPECCIÓN #{self.id_inspeccion} ---\n"
          f"Producto: {self.producto.get_nombre()}\n"
          f"ID Producto: {self.producto.get_id_producto()}\n"
          f"Proveedor: {self.proveedor.get_nombre()}\n"
          f"ID Proveedor: {self.proveedor.get_id_proveedor()}\n"
          f"Almacén: {self.almacen.get_nombre()}\n"
          f"ID Almacén: {self.almacen.get_id_almacen()}\n"
          f"Estado actual: {self.estado_actual}\n"
          f"Estado final: {self.estado_final}\n"
          f"Fecha estado actual: {self.fecha_estado_actual}\n"
          f"Cantidad devuelta: {self.cantidad_devuelta}\n"
          f"Observaciones: {self.observaciones}\n"
          f"Acción derivada: {self.determinar_accion()}\n"
      )
      if self.fecha_reingreso_almacen:
            resultado += f"Fecha reingreso a almacén: {self.fecha_reingreso_almacen}\n"
      if self.fecha_devolucion_proveedor:
            resultado += f"Fecha devolución al proveedor: {self.fecha_devolucion_proveedor}\n"
      return resultado
    def mostrar_historial(self):
        texto="--- HISTORIAL DE ESTADOS ---\n"
        for i, evento in enumerate(self.historial_estados, 1):
            texto += (f"{i}. Estado actual: {evento['estado_actual']}, "
                      f"Estado final: {evento['estado_final']}, "
                      f"Fecha: {evento['fecha']}\n")
        return texto
proveedor=Proveedor(16, "Angel Gallardo")
almacen=Almacen(1, "Almacén Central")
producto=Producto("H80230", "Martillo")
fecha_inicial=date(2025,6,2)
calidad=Calidad(
    id_inspeccion="HJQ1612",
    proveedor=proveedor,
    producto=producto,
    almacen=almacen,
    estado_actual="Recibido",
    estado_final="Pendiente",
    fecha_estado_actual=date.today(),
    cantidad_devuelta=0,
    observaciones="En revisión"
)
try:
  calidad.cambiar_estado("Autorizado para venta", date.today(), nuevo_estado_final="cerrado")
except ValueError:
  print(f"Error al cambiar estado. Revise lista de estados")
else:
  print("Estado cambiado con exito")
  print(calidad.mostrar_resultados())
  print(calidad.mostrar_historial())
```

```python
class Factura():
    def __init__(self, codigo_factura, productos, codigo_sucursal, codigo_cliente, numero_factura, iva_producto):
        self.codigo_factura = codigo_factura
        self.productos = productos
        self.codigo_sucursal = codigo_sucursal
        self.codigo_cliente = codigo_cliente
        self.numero_factura = numero_factura
        self.iva_producto = iva_producto
        self.total_factura = 0
        for producto in self.productos:
            self.total_factura += producto['precio_producto'] * producto['cantidad_producto'] * self.iva_producto

    def info(self):
        separador = "-------------------------------"
        info_productos = ""

        for producto in self.productos:
            info_productos += (
                f"\n  {producto['codigo_producto']}                |     $ {producto['precio_producto']}     |       {producto['cantidad_producto']}     |   {self.iva_producto}"
                f"\n{separador}----{separador}"
            )

        return (
            f"{separador} B {separador}"
            f"\n  FACTURA N°:           {self.numero_factura}        |  CODIGO DE CLIENTE:    {self.codigo_cliente}"
            f"\n  CODIGO DE FACTURA:    {self.codigo_factura}    |  CODIGO DE SUCURSAL:   {self.codigo_sucursal}"
            f"\n{separador}---{separador}"
            f"\n  CODIGO DE PRODUCTO:  |    PRECIO    |   CANTIDAD  |   IVA"
            f"{info_productos}"
            f"\n                                  TOTAL FACTURA:        {self.total_factura:.2f}"
            f"\n{separador}----{separador}"
        )

fk_producto = ["mtc01", "mtc02", "fjs01", "hst01"]
fk_facturas = ["fac01", "fc02"]

articulos = [
    {'codigo_producto': fk_producto[0],
     'precio_producto': 10,
     'cantidad_producto': 2
     },
    {'codigo_producto': fk_producto[1],
     'precio_producto': 15,
     'cantidad_producto': 1
     },
    {'codigo_producto': fk_producto[2],
     'precio_producto': 25,
     'cantidad_producto': 3
     },
    {'codigo_producto': fk_producto[3],
     'precio_producto': 30,
     'cantidad_producto': 4}
]

facturas = {
    #         codigo factura  ,    productos            , sucursal ,cliene, nfac, iva
    1: Factura(fk_facturas[0],[articulos[0], articulos[1]],"SC-1001","CL-01", 1, 1.21),
    2: Factura(fk_facturas[1],[articulos[2], articulos[3]],"SC-1001","CL-01", 2, 1.21)
}

for numero, factura in facturas.items():
    print(factura.info())
```
